```python
class EfficientHedge:
    def __init__(self, po):
        self.po = po
        self.N = 100 # time disrectization    
        self.S0 = 100 # initial value of the asset
        self.T = 10 # maturity
        self.dt = self.T / 100
        self.initial_wealth = 70

        if po == np.inf:
            def ploss(payoff, outputs):
                loss = tf.math.reduce_max(tf.nn.relu(payoff - outputs),keepdims = True)
                return loss
        elif po == 0:
            def ploss(payoff, outputs):
                loss = tf.math.sign(tf.nn.relu(payoff - outputs))
#                 loss = tf.nn.sigmoid(payoff - outputs)
                return loss
        else:
            def ploss(payoff, outputs):
                loss = (tf.nn.softplus(payoff - outputs))**po
                return loss
        self.ploss = ploss

    def prepare(self):        
        self.m = 1 # dimension of price
        self.mu = 0.08
        self.sigma = 0.3
        self.Ktrain = 10**5
        self.price_path, self.time_grid = utils_efficient.simulate_GBM(self.m,self.Ktrain,self.N,self.T,\
                                                             self.mu,self.sigma,self.S0, 'equi')
        self.payoff = np.ones_like(self.price_path[:,-1])*self.S0
        self.model_hedge, self.Network0, self.Networks = build_dynamic(self.m, self.N, False, False, self.initial_wealth, self.po, self.ploss)   
        self.xtrain = [self.price_path, self.payoff]  # input be price_path
        self.ytrain = self.payoff*0  # output be payoff     
        
    def train(self):
        def zeroloss(y_true, y_predict):
            return tf.reduce_sum(y_predict*0)
        optimizer = tf.keras.optimizers.Adam(learning_rate=0.01) # specify the optmizer 
        self.model_hedge.compile(optimizer = optimizer,loss=zeroloss) # specify the expected shortfall
        self.model_hedge.fit(x=self.xtrain,y=self.ytrain, epochs=10,verbose=True,batch_size=256) # train the model
    def predict(self):
        self.hedge_output = self.model_hedge.predict(self.xtrain) # compute the output (deep replicate payoff) with trained model
        print('ploss')
        print('train: ',tf.reduce_mean(self.ploss(self.payoff, self.hedge_output)).numpy())
        print('do nothing: ',tf.reduce_mean(self.ploss(self.payoff, self.initial_wealth)).numpy())

        plt.scatter(self.price_path[:,-1,0], self.hedge_output[:,0], s = 1, alpha = 0.5, label = 'deep hedge test ')   # deep replicate payoff 
```

```python
alpha = 100
beta = 10
def build_dynamic(m, N, control_path, trans_cost, initial_wealth, po, ploss):
    
    L = 3 # number of layers in strategy
    n = m + 20  # nodes in the first but last layers
    Networks = utils_efficient.build_network(m, n , L, N)
    Network0 = keras.layers.Dense(1, use_bias=False)

    price = keras.Input(shape=(N+1,m))   # S_{t}; t=0,..,N+1; (batch, N+1, m)
    payoff = keras.Input(shape=(1))
    inputs = [price, payoff]
    price_difference = price[:,1:,:] - price[:,:-1,:]  # dS_{t}; t=0,..,N; (batch, N, m)
#     premium = Network0(tf.ones_like(price[:,0,:1])) # premium; (batch, 1)
    premium = initial_wealth
    HEDGE = [None]*(N+1)
    HEDGE[0] = tf.zeros_like(price[:,0,:])
    STRATEGY = [None]*N
    ADMISSIBLE = tf.zeros_like(price[:,0,:])
    for j in range(N):
        I = tf.math.log(price[:,j,:])
        STRATEGY[j] = Networks[j](I) # H_{t} = nn(S_{t}); (batch, m)
        HEDGE[j+1] = HEDGE[j] + STRATEGY[j] * price_difference[:,j,:] # dX_{t} = H_{t}dS_{t}; (batch, m)
        ADMISSIBLE = tf.math.minimum(ADMISSIBLE, HEDGE[j+1] + premium)
    outputs = premium + tf.math.reduce_sum(HEDGE[-1],axis = -1, keepdims = True) # premium + \int_{0}^{T}H_{t}dS_{t}; (batch, m)    
    model_hedge = keras.Model(inputs = inputs, outputs=outputs)

# Define LOSS


    loss1 = tf.reduce_mean(ploss(payoff, outputs))
    model_hedge.add_loss(loss1) 
    model_hedge.add_metric(loss1, name='p-loss')

    
    return model_hedge, Network0, Networks
```

```python
efficient_hedge_1 = EfficientHedge(1)

efficient_hedge = efficient_hedge_1
efficient_hedge.prepare()
efficient_hedge.train()
efficient_hedge.predict()
```

```python
efficient_hedge = efficient_hedge_1
print('ploss')
print('train: ',tf.reduce_mean(efficient_hedge.ploss(efficient_hedge.payoff, efficient_hedge.hedge_output)).numpy())
print('do nothing: ',tf.reduce_mean(efficient_hedge.ploss(efficient_hedge.payoff, efficient_hedge.initial_wealth)).numpy())
plt.scatter(efficient_hedge.price_path[:,-1,0], efficient_hedge.hedge_output[:,0], s = 1, alpha = 0.1, label = 'deep hedge test ')   # deep replicate payoff 
plt.grid()
plt.xlim([0,500])
plt.show()
```
